```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('/content/prostate.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates
```

```python
data = df.drop('ID',axis = 1)
```

```python
data.head()
```

```python
x = df.drop('PSA',axis = 1)
y = df['PSA']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_test = (y_test>=10).astype(int)
```

```python
y_test
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred = (y_pred>=10).astype(int)

y_pred
```

```python
from sklearn.metrics import accuracy_score , precision_score , f1_score , recall_score
```

```python
print(accuracy_score(y_test,y_pred),f1_score(y_test,y_pred),precision_score(y_test,y_pred)
,recall_score(y_test,y_pred)
)
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test,y_pred)
print(score)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['PSA'],y, label='Observed Data', alpha=0.7)
plt.plot(df['PSA'],y, color='red', linewidth=1, label='Regression Line')
plt.title(f'Linear Regression Fit (R-squared: {score:.2f})')
plt.xlabel('y_Pred')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
```

```python
import pickle
filename = "prostate.pkl"
with open(filename, 'wb') as file:
    pickle.dump(model, file)
```
